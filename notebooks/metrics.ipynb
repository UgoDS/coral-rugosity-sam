```python
import pandas as pd
import seaborn as sns
from scipy import stats
from matplotlib import pyplot as plt
sns.set_theme() # Setting seaborn as default style even if use only matpl
```

```python
df = pd.read_csv("/Users/ugo/Documents/rugosity/data/RugoData_v0.csv")
df
```

```python
g = sns.lmplot(x="RugoCnT", y="RugoPh", data=df, col="Depth", row="Site")

def annotate(data, **kws):
    r, p = stats.pearsonr(data['RugoCnT'], data['RugoPh'])
    ax = plt.gca()
    ax.text(.05, .8, 'r={:.2f}, p={:.2g}'.format(r, p),
            transform=ax.transAxes)
    
g.map_dataframe(annotate)
plt.show()
```
